TRAVERSING THROUGH A DICTIONARY

```python
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}

def traverseDict(dict):
    for key in dict:
        print(key, dict[key])
traverseDict(myDict)
```

SEARCH FOR AN ELEMENT IN A DICTIONARY

```python
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}

def searcDict(dict, value):
    for key in dict:
        if dict[key] == value:
            return key, value
    return 'value does not exist'
searcDict(myDict, 26)
```

DICTIONARY METHODS

```python
#clear method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
myDict.clear()
print(myDict)
```

```python
#copy method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
dict = myDict.copy()

print(myDict)
print(dict)
```

```python
#fromkeys method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}

new_dict = {}.fromkeys([1,2,3], 0)
print(new_dict)

new_dict2 = {}.fromkeys([1,2,3])
print(new_dict2)
```

```python
#get method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.get('age', 27))
print(myDict.get('city', 27))
print(myDict.get(27))
```

```python
#items method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.items())
```

```python
#keys method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.keys())
```

```python
#popitem method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.popitem())
print(myDict)
```

```python
#set default method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.setdefault('name1', 'added'))
print(myDict)
```

```python
#pop method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.pop('name'))
print(myDict)
```

```python
#values
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}
print(myDict.values())
print(myDict.keys())
```

```python
#update method
myDict = {'name': 'eddy', 'age': 26, 'address': 'london'}

newDict = {'name1': 'eddy1', 'age1': 27, 'address1': 'london1'}

myDict.update(newDict)
print(myDict)
```

DICTIONARY OPERATIONS

```python
my_dict = {
    1: "one",
    2:"two",
    3:"three",
    4:"four",
    5:"five",
    6:"six"
}   

#values
print(3 in my_dict)
print("three" in my_dict.values())

#len()
print(len(my_dict))

#all() - sees if all keys are true
print(all(my_dict)) #all keys have to be string or not zero else you'll get false

#any() - sees if any keys are true
print(any(my_dict))

#sorted()
print(sorted(my_dict))
```
